The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

![](Nobel_Prize.png)

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

## Importing Libraries & Loading the Dataset

```python
import pandas as pd
import seaborn as sns
import numpy as np

nobel = pd.read_csv("data/nobel.csv")
```

## Extracting Top Values of (gender) and (birth country)

```python
top_gender = nobel['sex'].mode()[0]
top_country = nobel['birth_country'].mode()[0]
```

```python
print(top_gender)
print(top_country)
```

## Identify the decade with the highest ratio of US-born winners

```python
nobel['winUSA'] = nobel['birth_country'] == 'United States of America'
nobel['decade'] = (nobel['year'] // 10) * 10
```

```python
nobel.head()
```

```python
total_decade = nobel.groupby('decade').size()
total_usa = nobel[nobel['winUSA']].groupby('decade').size()
```

```python
ratio = total_usa / total_decade
max_decade_usa = ratio.idxmax()
```

```python
print(max_decade_usa)
```

## Find the decade and category with the highest proportion of female laureates

```python
nobel['winFemale'] = nobel['sex'] == 'Female'
group = nobel.groupby(['decade', 'category'])
```

```python
F_ratio = group['winFemale'].mean()
```

```python
max_index = F_ratio.idxmax()
max_female_dict = {max_index[0]: max_index[1]}
```

```python
print(max_female_dict)
```

## Find first woman to win a Nobel Prize

```python
winFemale = nobel[nobel['winFemale'] == True]
firstYear = winFemale['year'].min()
first_female = winFemale[winFemale['year'] == firstYear]
```

```python
first_woman_name = first_female.iloc[0]['full_name']
first_woman_category = first_female.iloc[0]['category']
print(first_woman_name)
print(first_woman_category)
```

## Determine repeat winners

```python
repeated = nobel['full_name'].value_counts()
repeat_list = repeated[repeated >= 2].index.tolist()
```

| `'Vict Descent'` | Victim's descent:<ul><li>`A` - Other Asian</li><li>`B` - Black</li><li>`C` - Chinese</li><li>`D` - Cambodian</li><li>`F` - Filipino</li><li>`G` - Guamanian</li><li>`H` - Hispanic/Latin/Mexican</li><li>`I` - American Indian/Alaskan Native</li><li>`J` - Japanese</li><li>`K` - Korean</li><li>`L` - Laotian</li><li>`O` - Other</li><li>`P` - Pacific Islander</li><li>`S` - Samoan</li><li>`U` - Hawaiian</li><li>`V` - Vietnamese</li><li>`W` - White</li><li>`X` - Unknown</li><li>`Z` - Asian Indian</li> |

| `'LOCATION'` | Street address of the crime. |

```python
# Re-run this cell
#Â Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv("crimes.csv", dtype={"TIME OCC": str})
crimes.head()
```

## Finding the frequencies of crimes by the hour of occurrence

```python
crimes["Hour"] = crimes["TIME OCC"].str.zfill(4).str[:2].astype(int)
```

```python
peak_crime_hour = crimes["Hour"].value_counts().idxmax()
print(peak_crime_hour)
```

```python
crimes["Hour"].head()
```

```python
g = sns.catplot(data=crimes, x="Hour", kind="count", height=6, aspect=2)

g.fig.suptitle("Crime Frequency by Hour", y=1.03)
g.set_axis_labels("Hour of Day", "Number of Crimes")

plt.show()
```

## Identifying the area with the most night crime

```python
crimes['Area Night'] = np.nan
night_condition = (crimes['Hour'] >= 22) | (crimes['Hour'] <= 4)
crimes.loc[night_condition, 'Area Night'] = crimes.loc[night_condition, 'AREA NAME']
```

```python
peak_night_crime_location = crimes["Area Night"].value_counts().idxmax()
print(peak_night_crime_location)
```

```python
area_counts = crimes['Area Night'].value_counts()

plt.figure(figsize=(12, 6))
sns.lineplot(
    x=area_counts.index,
    y=area_counts.values,
    marker='o'
)

plt.title("Crime Frequency by Area (10PMâ€“3:59AM)", fontsize=14)
plt.xlabel("Area Name")
plt.ylabel("Number of Crimes")
plt.xticks(rotation=90)

plt.show()
```

## Crimes by age group

```python
bins = [0, 17, 25, 34, 44, 54, 64, np.inf]
labels = ["0-17", "18-25", "26-34", "35-44", "45-54", "55-64", "65+"]
```

```python
age_group = pd.cut(crimes["Vict Age"], bins=bins, labels=labels, right=True, include_lowest=True)
victim_ages = age_group.value_counts().sort_index()
```

```python
print(victim_ages)
```

```python
victim_ages.plot(kind="pie", figsize=(10, 6))

plt.title("Number of Crimes by Victim Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Crimes")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
